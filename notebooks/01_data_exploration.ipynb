# 01 — Data Exploration

Quick interactive look at the merged corpus before preprocessing.

Loads `data/raw/merged_corpus.jsonl`, displays per-source counts,
topic distribution, length histograms, and a peek at the YouTube,
synthetic, and existing-corpus samples.

```python
%load_ext autoreload
%autoreload 2

import json
from collections import Counter
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

from src.utils import project_root
```

```python
raw_path = project_root() / 'data/raw/merged_corpus.jsonl'
records = [json.loads(l) for l in raw_path.open('r', encoding='utf-8') if l.strip()]
df = pd.DataFrame(records)
df.head()
```

```python
df['source'].value_counts().plot(kind='bar', title='Records per source')
plt.show()
```

```python
df['topic'].value_counts().plot(kind='bar', title='Records per topic')
plt.show()
```

```python
for src, sub in df.groupby('source'):
    print('---', src, '---')
    for row in sub.head(3).itertuples():
        print('Q:', row.question[:200])
        print('A:', (row.answer or '')[:200])
        print()
```
