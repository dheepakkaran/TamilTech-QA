# 03 — Training Monitoring

Visualise training-loss curves and sample generations produced by
`SampleGenerationCallback` while training is still running.

```python
import json
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

from src.utils import project_root

root = project_root() / 'outputs'
```

```python
for hist in root.glob('*/loss_history.json'):
    data = json.loads(hist.read_text())
    df = pd.DataFrame(data)
    if 'loss' in df.columns:
        plt.plot(df['step'], df['loss'], label=hist.parent.name)
plt.legend(); plt.xlabel('step'); plt.ylabel('loss'); plt.title('Training loss')
plt.show()
```

```python
for samp in root.glob('*/sample_outputs.jsonl'):
    print('===', samp.parent.name, '===')
    lines = samp.read_text(encoding='utf-8').splitlines()[-5:]
    for l in lines:
        rec = json.loads(l)
        print('step', rec['step'])
        print('Q:', rec['prompt'][:120])
        print('A:', rec['completion'][:240])
        print()
```
