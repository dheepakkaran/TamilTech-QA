# 02 — Preprocessing Analysis

Inspect the funnel that the cleaning + Tanglish-band filter creates,
and visualise the `tanglish_ratio` distribution in the final dataset.

```python
import json
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

from src.utils import project_root

root = project_root()
```

```python
stats = json.loads((root / 'data/processed/clean_stats.json').read_text())
pd.Series(stats).plot(kind='bar', title='Cleaning funnel')
plt.show()
```

```python
dfs = {}
for split in ('train', 'val', 'test'):
    p = root / f'data/final/{split}.jsonl'
    dfs[split] = pd.DataFrame(json.loads(l) for l in p.open('r', encoding='utf-8') if l.strip())
all_df = pd.concat(dfs.values())
all_df['tanglish_ratio'].hist(bins=30)
plt.axvline(0.15, color='r', ls='--'); plt.axvline(0.85, color='r', ls='--')
plt.title('Tanglish ratio — final dataset')
plt.show()
```

```python
(all_df.groupby('topic').size().sort_values(ascending=False).plot(kind='bar', title='Final corpus topic breakdown'))
plt.show()
```
